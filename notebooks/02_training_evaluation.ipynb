```python
# Cell 3: LOOCV sweep on the 46-patient cohort WITHOUT softplus scaling (no scale parameter)
import os
import sys
import warnings
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
from concurrent.futures import ProcessPoolExecutor
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from scipy import stats
from src.data.dataset import ClinicalAttentionDataset

warnings.filterwarnings('ignore')

class CAFNoScale(nn.Module):
    def __init__(self):
        super().__init__()
        self.gating = nn.Linear(3, 3)
        self.classifier_bias = nn.Parameter(torch.zeros(1))

    def forward(self, x_gate, s_tab, s_soft, s_hard):
        x_gate_vaping = x_gate[:, 2:5]
        gate_logits = self.gating(x_gate_vaping)
        probs = F.softmax(gate_logits, dim=1)
        p_tab = probs[:, 0:1]
        p_soft = probs[:, 1:2]
        p_hard = probs[:, 2:3]
        
        w_tab = p_tab * s_tab
        w_soft = p_soft * s_soft
        w_hard = p_hard * s_hard
        
        # Directly sum modalities plus bias (no scaling multiplier)
        logit = (w_tab + w_soft + w_hard) + self.classifier_bias
        return logit, probs

def evaluate_single_seed_noscale(seed, folds_data):
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    
    y_caf_pred, y_caf_prob = [], []
    caf_attention_weights = []
    
    for fold in folds_data:
        s_tab_train = fold['s_tab_train']
        s_tab_test = fold['s_tab_test']
        s_soft_train = fold['s_soft_train']
        s_soft_test = fold['s_soft_test']
        s_hard_train = fold['s_hard_train']
        s_hard_test = fold['s_hard_test']
        y_train = fold['y_train']
        X_gate_train = fold['X_gate_train']
        X_gate_test = fold['X_gate_test']
        
        model_caf = CAFNoScale()
        caf_criterion = nn.BCEWithLogitsLoss()
        caf_optimizer = optim.LBFGS(
            model_caf.parameters(), 
            lr=0.05, 
            max_iter=100,
            tolerance_grad=1e-7,
            tolerance_change=1e-9,
            history_size=50
        )
        
        s_tab_train_t = torch.tensor(s_tab_train, dtype=torch.float32).unsqueeze(1)
        s_soft_train_t = torch.tensor(s_soft_train, dtype=torch.float32).unsqueeze(1)
        s_hard_train_t = torch.tensor(s_hard_train, dtype=torch.float32).unsqueeze(1)
        y_train_t = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
        X_gate_train_t = torch.tensor(X_gate_train, dtype=torch.float32)
        X_gate_test_t = torch.tensor(X_gate_test, dtype=torch.float32)
        
        def caf_closure():
            caf_optimizer.zero_grad()
            logits, _ = model_caf(X_gate_train_t, s_tab_train_t, s_soft_train_t, s_hard_train_t)
            loss = caf_criterion(logits, y_train_t)
            loss.backward()
            return loss
            
        model_caf.train()
        caf_optimizer.step(caf_closure)
        
        s_tab_test_t = torch.tensor(s_tab_test, dtype=torch.float32).unsqueeze(0)
        s_soft_test_t = torch.tensor(s_soft_test, dtype=torch.float32).unsqueeze(0)
        s_hard_test_t = torch.tensor(s_hard_test, dtype=torch.float32).unsqueeze(0)
        model_caf.eval()
        with torch.no_grad():
            logit_caf, probs_caf = model_caf(X_gate_test_t, s_tab_test_t, s_soft_test_t, s_hard_test_t)
            prob_caf = torch.sigmoid(logit_caf).item()
            y_caf_pred.append(1 if prob_caf >= 0.5 else 0)
            y_caf_prob.append(prob_caf)
            caf_attention_weights.append(probs_caf.numpy().flatten())
            
    return np.array(y_caf_pred), np.array(y_caf_prob), np.array(caf_attention_weights)

# Load data & setup folds
cleaned_xlsx_path = os.path.join(project_root, 'data', 'questionnaire_cleaned.xlsx')
dataset = ClinicalAttentionDataset(cleaned_xlsx_path)

exclude_ids = {6, 7, 31, 34}
keep_indices = [i for i, pid in enumerate(dataset.patient_ids) if pid not in exclude_ids]

dataset.patient_ids = [dataset.patient_ids[i] for i in keep_indices]
dataset.labels = [dataset.labels[i] for i in keep_indices]
dataset.X_gate = dataset.X_gate[keep_indices]
dataset.s_soft = dataset.s_soft[keep_indices]
dataset.s_hard = dataset.s_hard[keep_indices]
dataset.X_tab = dataset.X_tab[keep_indices]

y_labels = np.array(dataset.labels)
X_gate_all = dataset.X_gate
s_soft_all = dataset.s_soft
s_hard_all = dataset.s_hard
X_tab_all = dataset.X_tab

loo = LeaveOneOut()
folds_data = []

for train_idx, test_idx in loo.split(X_tab_all):
    X_train_tab, X_test_tab = X_tab_all[train_idx], X_tab_all[test_idx]
    y_train, y_test = y_labels[train_idx], y_labels[test_idx]
    
    selector = SelectKBest(score_func=f_classif, k=16)
    X_train_sel = selector.fit_transform(X_train_tab, y_train)
    X_test_sel = selector.transform(X_test_tab)
    
    clf_tab = LogisticRegression(C=1.0, solver='liblinear', random_state=42)
    clf_tab.fit(X_train_sel, y_train)
    
    s_tab_train = clf_tab.predict_proba(X_train_sel)[:, 1]
    s_tab_test = clf_tab.predict_proba(X_test_sel)[:, 1]
    
    folds_data.append({
        's_tab_train': s_tab_train,
        's_tab_test': s_tab_test,
        's_soft_train': s_soft_all[train_idx],
        's_soft_test': s_soft_all[test_idx],
        's_hard_train': s_hard_all[train_idx],
        's_hard_test': s_hard_all[test_idx],
        'y_train': y_train,
        'y_test': y_test,
        'X_gate_train': X_gate_all[train_idx],
        'X_gate_test': X_gate_all[test_idx]
    })

seeds = list(range(100))
results = []

print("Running 100 seeds in parallel for CAF No-Scale (N=46)... ")

max_workers = min(os.cpu_count() or 4, 8)
with ProcessPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(evaluate_single_seed_noscale, seed, folds_data) for seed in seeds]
    for idx, future in enumerate(futures):
        results.append(future.result())

accs, precs, recs, f1s, aucs = [], [], [], [], []
is_vaper = (X_gate_all[:, 2] > 0.0).astype(int)
vaper_indices = np.where(is_vaper == 1)[0]
non_vaper_indices = np.where(is_vaper == 0)[0]

p_values = []
p_soft_non_vaper_means = []
p_soft_vaper_means = []

for pred, prob, weights in results:
    accs.append(accuracy_score(y_labels, pred))
    precs.append(precision_score(y_labels, pred, zero_division=0))
    recs.append(recall_score(y_labels, pred, zero_division=0))
    f1s.append(f1_score(y_labels, pred, zero_division=0))
    aucs.append(roc_auc_score(y_labels, prob))
    
    p_soft_w = weights[:, 1]
    p_soft_non_vaper_means.append(p_soft_w[non_vaper_indices].mean())
    p_soft_vaper_means.append(p_soft_w[vaper_indices].mean())
    
    _, p_val = stats.ttest_ind(p_soft_w[vaper_indices], p_soft_w[non_vaper_indices], equal_var=False)
    p_values.append(p_val)

print("\n" + "=" * 96)
print("LOOCV MULTI-SEED BENCHMARK FOR CAF NO-SCALE (N=46)")
print("=" * 96)
print(f"Accuracy:  {np.mean(accs)*100:.1f}±{np.std(accs)*100:.1f}%")
print(f"Precision: {np.mean(precs)*100:.1f}±{np.std(precs)*100:.1f}%")
print(f"Recall:    {np.mean(recs)*100:.1f}±{np.std(recs)*100:.1f}%")
print(f"F1-Score:  {np.mean(f1s):.3f}±{np.std(f1s):.3f}")
print(f"ROC-AUC:   {np.mean(aucs):.3f}±{np.std(aucs):.3f}")
print("-" * 96)
print("Gating weights:")
print(f"  Non-Vaper P_soft:  {np.mean(p_soft_non_vaper_means):.4f}±{np.std(p_soft_non_vaper_means):.4f}")
print(f"  Active Vaper P_soft: {np.mean(p_soft_vaper_means):.4f}±{np.std(p_soft_vaper_means):.4f}")
print("-" * 96)
success_seeds = sum(1 for p in p_values if p < 0.05)
print(f"Clinical shift success rate (p < 0.05): {success_seeds / 100 * 100:.1f}%")
print(f"Median Welch's t-test p-value:         {np.median(p_values):.5e}")
print("=" * 96)
```

```python
# Cell 4: Full Comparative Benchmark of all 3 configurations (Fixed, Softplus, Unconstrained) across 100 seeds
import os
import sys
import warnings
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
from concurrent.futures import ProcessPoolExecutor
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
from scipy import stats
from src.data.dataset import ClinicalAttentionDataset

warnings.filterwarnings('ignore')

class CAFFixed(nn.Module):
    def __init__(self):
        super().__init__()
        self.gating = nn.Linear(3, 3)
        self.classifier_bias = nn.Parameter(torch.zeros(1))
    def forward(self, x_gate, s_tab, s_soft, s_hard):
        x_gate_vaping = x_gate[:, 2:5]
        gate_logits = self.gating(x_gate_vaping)
        probs = F.softmax(gate_logits, dim=1)
        w_tab = probs[:, 0:1] * s_tab
        w_soft = probs[:, 1:2] * s_soft
        w_hard = probs[:, 2:3] * s_hard
        logit = (w_tab + w_soft + w_hard) + self.classifier_bias
        return logit, probs

class CAFSoftplus(nn.Module):
    def __init__(self):
        super().__init__()
        self.gating = nn.Linear(3, 3)
        self.scale_param = nn.Parameter(torch.tensor([0.5413]))  # softplus(0.5413) approx 1.0
        self.classifier_bias = nn.Parameter(torch.zeros(1))
    def forward(self, x_gate, s_tab, s_soft, s_hard):
        x_gate_vaping = x_gate[:, 2:5]
        gate_logits = self.gating(x_gate_vaping)
        probs = F.softmax(gate_logits, dim=1)
        w_tab = probs[:, 0:1] * s_tab
        w_soft = probs[:, 1:2] * s_soft
        w_hard = probs[:, 2:3] * s_hard
        scale = F.softplus(self.scale_param)
        logit = scale * (w_tab + w_soft + w_hard) + self.classifier_bias
        return logit, probs

class CAFUnconstrained(nn.Module):
    def __init__(self):
        super().__init__()
        self.gating = nn.Linear(3, 3)
        self.scale = nn.Parameter(torch.tensor(2.0))
        self.classifier_bias = nn.Parameter(torch.zeros(1))
    def forward(self, x_gate, s_tab, s_soft, s_hard):
        x_gate_vaping = x_gate[:, 2:5]
        gate_logits = self.gating(x_gate_vaping)
        probs = F.softmax(gate_logits, dim=1)
        w_tab = probs[:, 0:1] * s_tab
        w_soft = probs[:, 1:2] * s_soft
        w_hard = probs[:, 2:3] * s_hard
        logit = self.scale * (w_tab + w_soft + w_hard) + self.classifier_bias
        return logit, probs

def train_and_eval_model(model_class, folds_data, seed):
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    
    y_pred, y_prob = [], []
    attention_weights = []
    scale_values = []
    
    for fold in folds_data:
        s_tab_train = fold['s_tab_train']
        s_tab_test = fold['s_tab_test']
        s_soft_train = fold['s_soft_train']
        s_soft_test = fold['s_soft_test']
        s_hard_train = fold['s_hard_train']
        s_hard_test = fold['s_hard_test']
        y_train = fold['y_train']
        X_gate_train = fold['X_gate_train']
        X_gate_test = fold['X_gate_test']
        
        model = model_class()
        criterion = nn.BCEWithLogitsLoss()
        optimizer = optim.LBFGS(
            model.parameters(), 
            lr=0.05, 
            max_iter=100,
            tolerance_grad=1e-7,
            tolerance_change=1e-9,
            history_size=50
        )
        
        s_tab_train_t = torch.tensor(s_tab_train, dtype=torch.float32).unsqueeze(1)
        s_soft_train_t = torch.tensor(s_soft_train, dtype=torch.float32).unsqueeze(1)
        s_hard_train_t = torch.tensor(s_hard_train, dtype=torch.float32).unsqueeze(1)
        y_train_t = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
        X_gate_train_t = torch.tensor(X_gate_train, dtype=torch.float32)
        X_gate_test_t = torch.tensor(X_gate_test, dtype=torch.float32)
        
        def closure():
            optimizer.zero_grad()
            logits, _ = model(X_gate_train_t, s_tab_train_t, s_soft_train_t, s_hard_train_t)
            loss = criterion(logits, y_train_t)
            loss.backward()
            return loss
            
        model.train()
        optimizer.step(closure)
        
        if hasattr(model, 'scale'):
            scale_values.append(model.scale.item())
        elif hasattr(model, 'scale_param'):
            scale_values.append(F.softplus(model.scale_param).item())
        else:
            scale_values.append(1.0)
            
        s_tab_test_t = torch.tensor(s_tab_test, dtype=torch.float32).unsqueeze(0)
        s_soft_test_t = torch.tensor(s_soft_test, dtype=torch.float32).unsqueeze(0)
        s_hard_test_t = torch.tensor(s_hard_test, dtype=torch.float32).unsqueeze(0)
        
        model.eval()
        with torch.no_grad():
            logit, probs = model(X_gate_test_t, s_tab_test_t, s_soft_test_t, s_hard_test_t)
            prob = torch.sigmoid(logit).item()
            y_pred.append(1 if prob >= 0.5 else 0)
            y_prob.append(prob)
            attention_weights.append(probs.numpy().flatten())
            
    return np.array(y_pred), np.array(y_prob), np.array(attention_weights), np.array(scale_values)

def evaluate_single_seed_all(seed, folds_data):
    pred_a, prob_a, weights_a, scales_a = train_and_eval_model(CAFFixed, folds_data, seed)
    pred_b, prob_b, weights_b, scales_b = train_and_eval_model(CAFSoftplus, folds_data, seed)
    pred_c, prob_c, weights_c, scales_c = train_and_eval_model(CAFUnconstrained, folds_data, seed)
    return {
        'fixed': (pred_a, prob_a, weights_a, scales_a),
        'softplus': (pred_b, prob_b, weights_b, scales_b),
        'unconstrained': (pred_c, prob_c, weights_c, scales_c)
    }

# Set up data
cleaned_xlsx_path = os.path.join(project_root, 'data', 'questionnaire_cleaned.xlsx')
dataset = ClinicalAttentionDataset(cleaned_xlsx_path)

exclude_ids = {6, 7, 31, 34}
keep_indices = [i for i, pid in enumerate(dataset.patient_ids) if pid not in exclude_ids]

dataset.patient_ids = [dataset.patient_ids[i] for i in keep_indices]
dataset.labels = [dataset.labels[i] for i in keep_indices]
dataset.X_gate = dataset.X_gate[keep_indices]
dataset.s_soft = dataset.s_soft[keep_indices]
dataset.s_hard = dataset.s_hard[keep_indices]
dataset.X_tab = dataset.X_tab[keep_indices]

y_labels = np.array(dataset.labels)
X_gate_all = dataset.X_gate
s_soft_all = dataset.s_soft
s_hard_all = dataset.s_hard
X_tab_all = dataset.X_tab

loo = LeaveOneOut()
folds_data = []
for train_idx, test_idx in loo.split(X_tab_all):
    X_train_tab, X_test_tab = X_tab_all[train_idx], X_tab_all[test_idx]
    y_train = y_labels[train_idx]
    
    selector = SelectKBest(score_func=f_classif, k=16)
    X_train_sel = selector.fit_transform(X_train_tab, y_train)
    X_test_sel = selector.transform(X_test_tab)
    
    clf_tab = LogisticRegression(C=1.0, solver='liblinear', random_state=42)
    clf_tab.fit(X_train_sel, y_train)
    
    s_tab_train = clf_tab.predict_proba(X_train_sel)[:, 1]
    s_tab_test = clf_tab.predict_proba(X_test_sel)[:, 1]
    
    folds_data.append({
        's_tab_train': s_tab_train,
        's_tab_test': s_tab_test,
        's_soft_train': s_soft_all[train_idx],
        's_soft_test': s_soft_all[test_idx],
        's_hard_train': s_hard_all[train_idx],
        's_hard_test': s_hard_all[test_idx],
        'y_train': y_train,
        'X_gate_train': X_gate_all[train_idx],
        'X_gate_test': X_gate_all[test_idx]
    })

seeds = list(range(100))
results = []

print("Running 100 seeds for comparison...")
max_workers = min(os.cpu_count() or 4, 8)
with ProcessPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(evaluate_single_seed_all, seed, folds_data) for seed in seeds]
    for idx, future in enumerate(futures):
        results.append(future.result())

is_vaper = (X_gate_all[:, 2] > 0.0).astype(int)
vaper_indices = np.where(is_vaper == 1)[0]
non_vaper_indices = np.where(is_vaper == 0)[0]

metrics = {
    'fixed': {'acc': [], 'auc': [], 'p_vals': [], 'p_soft_nv': [], 'p_soft_v': []},
    'softplus': {'acc': [], 'auc': [], 'p_vals': [], 'p_soft_nv': [], 'p_soft_v': [], 'scales': []},
    'unconstrained': {'acc': [], 'auc': [], 'p_vals': [], 'p_soft_nv': [], 'p_soft_v': [], 'scales': []}
}

for r in results:
    for key in ['fixed', 'softplus', 'unconstrained']:
        pred, prob, weights, scales = r[key]
        metrics[key]['acc'].append(accuracy_score(y_labels, pred))
        metrics[key]['auc'].append(roc_auc_score(y_labels, prob))
        p_soft_w = weights[:, 1]
        metrics[key]['p_soft_nv'].append(p_soft_w[non_vaper_indices].mean())
        metrics[key]['p_soft_v'].append(p_soft_w[vaper_indices].mean())
        _, p_val = stats.ttest_ind(p_soft_w[vaper_indices], p_soft_w[non_vaper_indices], equal_var=False)
        metrics[key]['p_vals'].append(p_val)
        if 'scales' in metrics[key]:
            metrics[key]['scales'].extend(list(scales))

print("\n" + "=" * 110)
print("COMPARATIVE MULTI-SEED BENCHMARK (Averaged over 100 seeds)")
print("=" * 110)
print(f"{'Model Configuration':<30} | {'Accuracy':<13} | {'ROC-AUC':<13} | {'t-test success':<15} | {'Median p-value':<15}")
print("-" * 110)
for key, name in [('fixed', 'A) Fixed Scale (1.0)'), ('softplus', 'B) Softplus Scale'), ('unconstrained', 'C) Unconstrained Scale')]:
    acc_m, acc_s = np.mean(metrics[key]['acc']), np.std(metrics[key]['acc'])
    auc_m, auc_s = np.mean(metrics[key]['auc']), np.std(metrics[key]['auc'])
    success = sum(1 for p in metrics[key]['p_vals'] if p < 0.05) / 100 * 100
    med_p = np.median(metrics[key]['p_vals'])
    print(f"{name:<30} | {acc_m*100:.1f}±{acc_s*100:.1f}% | {auc_m:.3f}±{auc_s:.3f} | {success:.1f}% | {med_p:.5e}")
print("=" * 110)

uncon_scales = np.array(metrics['unconstrained']['scales'])
print("\n" + "=" * 110)
print("SCALE PARAMETER ANALYSIS FOR UNCONSTRAINED CONFIGURATION (Model C)")
print("=" * 110)
print(f"Mean Scale:             {uncon_scales.mean():.4f}")
print(f"Median Scale:           {np.median(uncon_scales):.4f}")
print(f"Min Scale:              {uncon_scales.min():.4f}")
print(f"Max Scale:              {uncon_scales.max():.4f}")
print(f"Folds with Scale > 0:   {sum(1 for s in uncon_scales if s > 0)} / {len(uncon_scales)}")
print(f"Folds with Scale < 0:   {sum(1 for s in uncon_scales if s < 0)} / {len(uncon_scales)}")
print("=" * 110)

print("\n" + "=" * 110)
print("ROUTING COMPARISON (Mean Attention Weights P_soft)")
print("=" * 110)
for key, name in [('fixed', 'A) Fixed Scale (1.0)'), ('softplus', 'B) Softplus Scale'), ('unconstrained', 'C) Unconstrained Scale')]:
    nv_mean, nv_std = np.mean(metrics[key]['p_soft_nv']), np.std(metrics[key]['p_soft_nv'])
    v_mean, v_std = np.mean(metrics[key]['p_soft_v']), np.std(metrics[key]['p_soft_v'])
    print(f"{name:<30} | Non-Vaper P_soft: {nv_mean:.4f}±{nv_std:.4f} | Active Vaper P_soft: {v_mean:.4f}±{v_std:.4f}")
print("=" * 110)
```

```python
# Cell 5: Evaluation of Late Concatenation & Bilinear Fusion baselines with exhaustive grid-search tuning per fold
import os
import sys
import warnings
import numpy as np
import torch
from concurrent.futures import ProcessPoolExecutor
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score, roc_auc_score
from src.data.dataset import ClinicalAttentionDataset

warnings.filterwarnings('ignore')

def evaluate_single_seed_tuned(seed, folds_data):
    import random
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    
    y_concat_pred, y_concat_prob = [], []
    y_bilinear_pred, y_bilinear_prob = [], []
    
    for fold in folds_data:
        s_tab_train = fold['s_tab_train']
        s_tab_test = fold['s_tab_test']
        s_soft_train = fold['s_soft_train']
        s_soft_test = fold['s_soft_test']
        s_hard_train = fold['s_hard_train']
        s_hard_test = fold['s_hard_test']
        y_train = fold['y_train']
        y_test = fold['y_test']
        
        X_train_c = np.column_stack([s_tab_train, s_soft_train, s_hard_train])
        X_test_c = np.column_stack([s_tab_test, s_soft_test, s_hard_test])
        
        # Grid search over regularization strength C and penalty
        param_grid = {'C': [0.001, 0.01, 0.1, 0.5, 1.0, 5.0, 10.0, 50.0, 100.0], 'penalty': ['l1', 'l2']}
        clf_c = LogisticRegression(solver='liblinear', random_state=42)
        grid_c = GridSearchCV(clf_c, param_grid, cv=3, scoring='accuracy')
        grid_c.fit(X_train_c, y_train)
        best_clf_c = grid_c.best_estimator_
        y_concat_pred.append(best_clf_c.predict(X_test_c)[0])
        y_concat_prob.append(best_clf_c.predict_proba(X_test_c)[0, 1])
        
        def get_bilinear_terms(X):
            t = X[:, 0]
            s = X[:, 1]
            h = X[:, 2]
            return np.column_stack([t, s, h, t*s, t*h, s*h, t*s*h])
        X_train_b = get_bilinear_terms(X_train_c)
        X_test_b = get_bilinear_terms(X_test_c)
        
        clf_b = LogisticRegression(solver='liblinear', random_state=42)
        grid_b = GridSearchCV(clf_b, param_grid, cv=3, scoring='accuracy')
        grid_b.fit(X_train_b, y_train)
        best_clf_b = grid_b.best_estimator_
        y_bilinear_pred.append(best_clf_b.predict(X_test_b)[0])
        y_bilinear_prob.append(best_clf_b.predict_proba(X_test_b)[0, 1])
        
    return {
        'tuned_concat': (np.array(y_concat_pred), np.array(y_concat_prob)),
        'tuned_bilinear': (np.array(y_bilinear_pred), np.array(y_bilinear_prob))
    }

# Set up data
cleaned_xlsx_path = os.path.join(project_root, 'data', 'questionnaire_cleaned.xlsx')
dataset = ClinicalAttentionDataset(cleaned_xlsx_path)

exclude_ids = {6, 7, 31, 34}
keep_indices = [i for i, pid in enumerate(dataset.patient_ids) if pid not in exclude_ids]

dataset.patient_ids = [dataset.patient_ids[i] for i in keep_indices]
dataset.labels = [dataset.labels[i] for i in keep_indices]
dataset.X_gate = dataset.X_gate[keep_indices]
dataset.s_soft = dataset.s_soft[keep_indices]
dataset.s_hard = dataset.s_hard[keep_indices]
dataset.X_tab = dataset.X_tab[keep_indices]

y_labels = np.array(dataset.labels)
X_gate_all = dataset.X_gate
s_soft_all = dataset.s_soft
s_hard_all = dataset.s_hard
X_tab_all = dataset.X_tab

loo = LeaveOneOut()
folds_data = []
for train_idx, test_idx in loo.split(X_tab_all):
    X_train_tab, X_test_tab = X_tab_all[train_idx], X_tab_all[test_idx]
    y_train = y_labels[train_idx]
    
    selector = SelectKBest(score_func=f_classif, k=16)
    X_train_sel = selector.fit_transform(X_train_tab, y_train)
    X_test_sel = selector.transform(X_test_tab)
    
    clf_tab = LogisticRegression(C=1.0, solver='liblinear', random_state=42)
    clf_tab.fit(X_train_sel, y_train)
    
    s_tab_train = clf_tab.predict_proba(X_train_sel)[:, 1]
    s_tab_test = clf_tab.predict_proba(X_test_sel)[:, 1]
    
    folds_data.append({
        's_tab_train': s_tab_train,
        's_tab_test': s_tab_test,
        's_soft_train': s_soft_all[train_idx],
        's_soft_test': s_soft_all[test_idx],
        's_hard_train': s_hard_all[train_idx],
        's_hard_test': s_hard_all[test_idx],
        'y_train': y_train,
        'y_test': y_labels[test_idx]
    })

seeds = list(range(100))
results = []
print("Running 100 seeds for fully tuned baselines...")
max_workers = min(os.cpu_count() or 4, 8)
with ProcessPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(evaluate_single_seed_tuned, seed, folds_data) for seed in seeds]
    for idx, future in enumerate(futures):
        results.append(future.result())

metrics = {
    'tuned_concat': {'acc': [], 'auc': []},
    'tuned_bilinear': {'acc': [], 'auc': []}
}

for r in results:
    for key in ['tuned_concat', 'tuned_bilinear']:
        pred, prob = r[key]
        metrics[key]['acc'].append(accuracy_score(y_labels, pred))
        metrics[key]['auc'].append(roc_auc_score(y_labels, prob))

print("\n" + "=" * 96)
print("LOOCV MULTI-SEED BENCHMARK FOR TUNED BASELINE MODELS (100 Seeds, N=46)")
print("=" * 96)
print(f"{'Model Name':<35} | {'Accuracy':<15} | {'ROC-AUC':<15}")
print("-" * 96)
for key, name in [('tuned_concat', 'Tuned Late Concatenation Fusion'), ('tuned_bilinear', 'Tuned Bilinear Tensor Fusion')]:
    acc_m, acc_s = np.mean(metrics[key]['acc']), np.std(metrics[key]['acc'])
    auc_m, auc_s = np.mean(metrics[key]['auc']), np.std(metrics[key]['auc'])
    print(f"{name:<35} | {acc_m*100:.2f}±{acc_s*100:.2f}% | {auc_m:.4f}±{auc_s:.4f}")
print("=" * 96)
```

```python
# Cell 7: Multi-Seed Evaluation of CAF (Softplus scaling, N=46) without the Radiographic/X-ray (s_hard) Modality
import os
import sys
import warnings
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
from concurrent.futures import ProcessPoolExecutor
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score, roc_auc_score
from scipy import stats
from src.data.dataset import ClinicalAttentionDataset

warnings.filterwarnings('ignore')

class CAFSoftplusNoXray(nn.Module):
    def __init__(self):
        super().__init__()
        self.gating = nn.Linear(3, 2)
        self.scale_param = nn.Parameter(torch.tensor([0.5413]))  # softplus(0.5413) approx 1.0
        self.classifier_bias = nn.Parameter(torch.zeros(1))
        
    def forward(self, x_gate, s_tab, s_soft):
        x_gate_vaping = x_gate[:, 2:5]
        gate_logits = self.gating(x_gate_vaping)
        probs = F.softmax(gate_logits, dim=1)  # [Batch, 2]
        
        p_tab = probs[:, 0:1]
        p_soft = probs[:, 1:2]
        
        w_tab = p_tab * s_tab
        w_soft = p_soft * s_soft
        
        scale = F.softplus(self.scale_param)
        logit = scale * (w_tab + w_soft) + self.classifier_bias
        return logit, probs

def train_and_eval_no_xray(folds_data, seed):
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    
    y_pred, y_prob = [], []
    attention_weights = []
    
    for fold in folds_data:
        s_tab_train = fold['s_tab_train']
        s_tab_test = fold['s_tab_test']
        s_soft_train = fold['s_soft_train']
        s_soft_test = fold['s_soft_test']
        y_train = fold['y_train']
        X_gate_train = fold['X_gate_train']
        X_gate_test = fold['X_gate_test']
        
        model = CAFSoftplusNoXray()
        criterion = nn.BCEWithLogitsLoss()
        optimizer = optim.LBFGS(
            model.parameters(), 
            lr=0.05, 
            max_iter=100,
            tolerance_grad=1e-7,
            tolerance_change=1e-9,
            history_size=50
        )
        
        s_tab_train_t = torch.tensor(s_tab_train, dtype=torch.float32).unsqueeze(1)
        s_soft_train_t = torch.tensor(s_soft_train, dtype=torch.float32).unsqueeze(1)
        y_train_t = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
        X_gate_train_t = torch.tensor(X_gate_train, dtype=torch.float32)
        X_gate_test_t = torch.tensor(X_gate_test, dtype=torch.float32)
        
        def closure():
            optimizer.zero_grad()
            logits, _ = model(X_gate_train_t, s_tab_train_t, s_soft_train_t)
            loss = criterion(logits, y_train_t)
            loss.backward()
            return loss
            
        model.train()
        optimizer.step(closure)
        
        s_tab_test_t = torch.tensor(s_tab_test, dtype=torch.float32).unsqueeze(0)
        s_soft_test_t = torch.tensor(s_soft_test, dtype=torch.float32).unsqueeze(0)
        
        model.eval()
        with torch.no_grad():
            logit, probs = model(X_gate_test_t, s_tab_test_t, s_soft_test_t)
            prob = torch.sigmoid(logit).item()
            y_pred.append(1 if prob >= 0.5 else 0)
            y_prob.append(prob)
            attention_weights.append(probs.numpy().flatten())
            
    return np.array(y_pred), np.array(y_prob), np.array(attention_weights)

project_root = r"c:\Users\abura\Development\CAF"
cleaned_xlsx_path = os.path.join(project_root, 'data', 'questionnaire_cleaned.xlsx')
dataset = ClinicalAttentionDataset(cleaned_xlsx_path)

exclude_ids = {6, 7, 31, 34}
keep_indices = [i for i, pid in enumerate(dataset.patient_ids) if pid not in exclude_ids]
dataset.patient_ids = [dataset.patient_ids[i] for i in keep_indices]
dataset.labels = [dataset.labels[i] for i in keep_indices]
dataset.X_gate = dataset.X_gate[keep_indices]
dataset.s_soft = dataset.s_soft[keep_indices]
dataset.X_tab = dataset.X_tab[keep_indices]

y_labels = np.array(dataset.labels)
X_gate_all = dataset.X_gate
s_soft_all = dataset.s_soft
X_tab_all = dataset.X_tab

loo = LeaveOneOut()
folds_data = []
for train_idx, test_idx in loo.split(X_tab_all):
    X_train_tab, X_test_tab = X_tab_all[train_idx], X_tab_all[test_idx]
    y_train = y_labels[train_idx]
    
    selector = SelectKBest(score_func=f_classif, k=16)
    X_train_sel = selector.fit_transform(X_train_tab, y_train)
    X_test_sel = selector.transform(X_test_tab)
    
    clf_tab = LogisticRegression(C=1.0, solver='liblinear', random_state=42)
    clf_tab.fit(X_train_sel, y_train)
    s_tab_train = clf_tab.predict_proba(X_train_sel)[:, 1]
    s_tab_test = clf_tab.predict_proba(X_test_sel)[:, 1]
    
    folds_data.append({
        's_tab_train': s_tab_train,
        's_tab_test': s_tab_test,
        's_soft_train': s_soft_all[train_idx],
        's_soft_test': s_soft_all[test_idx],
        'y_train': y_train,
        'X_gate_train': X_gate_all[train_idx],
        'X_gate_test': X_gate_all[test_idx]
    })

seeds = list(range(100))
results = []
print("Running 100 seeds for No-Xray CAF model...")
max_workers = min(os.cpu_count() or 4, 8)
with ProcessPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(train_and_eval_no_xray, folds_data, seed) for seed in seeds]
    for idx, future in enumerate(futures):
        results.append(future.result())

is_vaper = (X_gate_all[:, 2] > 0.0).astype(int)
vaper_indices = np.where(is_vaper == 1)[0]
non_vaper_indices = np.where(is_vaper == 0)[0]

accs, aucs, p_vals, p_soft_nv, p_soft_v = [], [], [], [], []
for r in results:
    pred, prob, weights = r
    accs.append(accuracy_score(y_labels, pred))
    aucs.append(roc_auc_score(y_labels, prob))
    
    p_soft_w = weights[:, 1]
    p_soft_nv.append(p_soft_w[non_vaper_indices].mean())
    p_soft_v.append(p_soft_w[vaper_indices].mean())
    
    _, p_val = stats.ttest_ind(p_soft_w[vaper_indices], p_soft_w[non_vaper_indices], equal_var=False)
    p_vals.append(p_val)

acc_m, acc_s = np.mean(accs), np.std(accs)
auc_m, auc_s = np.mean(aucs), np.std(aucs)
success = sum(1 for p in p_vals if p < 0.05) / 100 * 100
med_p = np.median(p_vals)
nv_mean = np.mean(p_soft_nv)
v_mean = np.mean(p_soft_v)

print("\n" + "=" * 96)
print("LOOCV MULTI-SEED EVALUATION: NO X-RAY MODALITY (100 Seeds, N=46)")
print("=" * 96)
print(f"Accuracy:        {acc_m*100:.2f}% ± {acc_s*100:.2f}%")
print(f"ROC-AUC:         {auc_m:.4f} ± {auc_s:.4f}")
print(f"t-test Success:  {success:.1f}% (seeds with p < 0.05)")
print(f"Median p-value:  {med_p:.5e}")
print(f"Non-Vaper P_soft:{nv_mean:.4f}")
print(f"Vaper P_soft:    {v_mean:.4f}")
print("=" * 96)
```

```python
# Cell 9: Evaluation of Late Concatenation & Bilinear Fusion using non-linear MLPs with activation sweeps (ReLU, LeakyReLU, GELU, Tanh)
import os
import sys
import warnings
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
from concurrent.futures import ProcessPoolExecutor
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import LeaveOneOut, KFold
from sklearn.metrics import accuracy_score, roc_auc_score
from src.data.dataset import ClinicalAttentionDataset

warnings.filterwarnings('ignore')

class MLPBaseline(nn.Module):
    def __init__(self, input_dim, hidden_dim, activation_cls):
        super().__init__()
        self.net = nn.Sequential(
            nn.Linear(input_dim, hidden_dim),
            activation_cls(),
            nn.Linear(hidden_dim, 1)
        )
    def forward(self, x):
        return self.net(x)

def train_and_test_mlp(X_train, y_train, X_test, hidden_dim, activation_cls, seed):
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    
    input_dim = X_train.shape[1]
    model = MLPBaseline(input_dim, hidden_dim, activation_cls)
    criterion = nn.BCEWithLogitsLoss()
    optimizer = optim.LBFGS(
        model.parameters(), 
        lr=0.05, 
        max_iter=100,
        tolerance_grad=1e-7,
        tolerance_change=1e-9,
        history_size=50
    )
    
    X_train_t = torch.tensor(X_train, dtype=torch.float32)
    y_train_t = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
    X_test_t = torch.tensor(X_test, dtype=torch.float32).unsqueeze(0)
    
    def closure():
        optimizer.zero_grad()
        logits = model(X_train_t)
        loss = criterion(logits, y_train_t)
        loss.backward()
        return loss
        
    model.train()
    optimizer.step(closure)
    
    model.eval()
    with torch.no_grad():
        logit = model(X_test_t)
        prob = torch.sigmoid(logit).item()
        pred = 1 if prob >= 0.5 else 0
    return pred, prob

def tune_mlp_hyperparams(X_train, y_train, input_dim):
    kf = KFold(n_splits=3, shuffle=True, random_state=42)
    hidden_dims = [4, 8, 16]
    activations = {
        'relu': nn.ReLU,
        'leaky_relu': nn.LeakyReLU,
        'gelu': nn.GELU,
        'tanh': nn.Tanh
    }
    
    best_score = -1.0
    best_config = None
    for h in hidden_dims:
        for act_name, act_cls in activations.items():
            scores = []
            for train_idx, val_idx in kf.split(X_train):
                X_tr, X_val = X_train[train_idx], X_train[val_idx]
                y_tr, y_val = y_train[train_idx], y_train[val_idx]
                preds = []
                for idx in range(len(X_val)):
                    pred, _ = train_and_test_mlp(X_tr, y_tr, X_val[idx], h, act_cls, seed=42)
                    preds.append(pred)
                scores.append(accuracy_score(y_val, preds))
                
            mean_score = np.mean(scores)
            if mean_score > best_score:
                best_score = mean_score
                best_config = (h, act_cls, act_name)
    return best_config[0], best_config[1], best_config[2]

def precompute_best_mlp_params(folds_data_c, folds_data_b):
    print("Precomputing best MLP configurations (hidden size & activation) per fold...")
    best_c_params = []
    best_b_params = []
    for fold in folds_data_c:
        s_tab_train = fold['s_tab_train']
        s_soft_train = fold['s_soft_train']
        s_hard_train = fold['s_hard_train']
        X_train_c = np.column_stack([s_tab_train, s_soft_train, s_hard_train])
        h, act_cls, act_name = tune_mlp_hyperparams(X_train_c, fold['y_train'], input_dim=3)
        best_c_params.append({'h': h, 'act_cls': act_cls, 'act_name': act_name})
        
    for fold in folds_data_b:
        s_tab_train = fold['s_tab_train']
        s_soft_train = fold['s_soft_train']
        s_hard_train = fold['s_hard_train']
        X_train_c = np.column_stack([s_tab_train, s_soft_train, s_hard_train])
        def get_bilinear_terms(X):
            t, s, h = X[:, 0], X[:, 1], X[:, 2]
            return np.column_stack([t, s, h, t*s, t*h, s*h, t*s*h])
        X_train_b = get_bilinear_terms(X_train_c)
        h, act_cls, act_name = tune_mlp_hyperparams(X_train_b, fold['y_train'], input_dim=7)
        best_b_params.append({'h': h, 'act_cls': act_cls, 'act_name': act_name})
    print("Precomputation complete.")
    return best_c_params, best_b_params

def evaluate_mlp_seed(seed, folds_data_c, folds_data_b, best_c_params, best_b_params):
    y_concat_pred, y_concat_prob = [], []
    y_bilinear_pred, y_bilinear_prob = [], []
    
    for i, fold in enumerate(folds_data_c):
        s_tab_train, s_tab_test = fold['s_tab_train'], fold['s_tab_test']
        s_soft_train, s_soft_test = fold['s_soft_train'], fold['s_soft_test']
        s_hard_train, s_hard_test = fold['s_hard_train'], fold['s_hard_test']
        X_train_c = np.column_stack([s_tab_train, s_soft_train, s_hard_train])
        X_test_c = np.column_stack([s_tab_test, s_soft_test, s_hard_test])
        
        cfg = best_c_params[i]
        pred, prob = train_and_test_mlp(X_train_c, fold['y_train'], X_test_c[0], cfg['h'], cfg['act_cls'], seed)
        y_concat_pred.append(pred)
        y_concat_prob.append(prob)
        
    for i, fold in enumerate(folds_data_b):
        s_tab_train, s_tab_test = fold['s_tab_train'], fold['s_tab_test']
        s_soft_train, s_soft_test = fold['s_soft_train'], fold['s_soft_test']
        s_hard_train, s_hard_test = fold['s_hard_train'], fold['s_hard_test']
        X_train_c = np.column_stack([s_tab_train, s_soft_train, s_hard_train])
        X_test_c = np.column_stack([s_tab_test, s_soft_test, s_hard_test])
        
        def get_bilinear_terms(X):
            t, s, h = X[:, 0], X[:, 1], X[:, 2]
            return np.column_stack([t, s, h, t*s, t*h, s*h, t*s*h])
        X_train_b = get_bilinear_terms(X_train_c)
        X_test_b = get_bilinear_terms(X_test_c)
        
        cfg = best_b_params[i]
        pred, prob = train_and_test_mlp(X_train_b, fold['y_train'], X_test_b[0], cfg['h'], cfg['act_cls'], seed)
        y_bilinear_pred.append(pred)
        y_bilinear_prob.append(prob)
        
    return {
        'concat': (np.array(y_concat_pred), np.array(y_concat_prob)),
        'bilinear': (np.array(y_bilinear_pred), np.array(y_bilinear_prob))
    }

project_root = r"c:\Users\abura\Development\CAF"
cleaned_xlsx_path = os.path.join(project_root, 'data', 'questionnaire_cleaned.xlsx')
dataset = ClinicalAttentionDataset(cleaned_xlsx_path)

exclude_ids = {6, 7, 31, 34}
keep_indices = [i for i, pid in enumerate(dataset.patient_ids) if pid not in exclude_ids]
dataset.patient_ids = [dataset.patient_ids[i] for i in keep_indices]
dataset.labels = [dataset.labels[i] for i in keep_indices]
dataset.s_soft = dataset.s_soft[keep_indices]
dataset.s_hard = dataset.s_hard[keep_indices]
dataset.X_tab = dataset.X_tab[keep_indices]

y_labels = np.array(dataset.labels)
s_soft_all = dataset.s_soft
s_hard_all = dataset.s_hard
X_tab_all = dataset.X_tab

loo = LeaveOneOut()
folds_data = []
for train_idx, test_idx in loo.split(X_tab_all):
    X_train_tab, X_test_tab = X_tab_all[train_idx], X_tab_all[test_idx]
    y_train = y_labels[train_idx]
    
    selector = SelectKBest(score_func=f_classif, k=16)
    X_train_sel = selector.fit_transform(X_train_tab, y_train)
    X_test_sel = selector.transform(X_test_tab)
    
    clf_tab = LogisticRegression(C=1.0, solver='liblinear', random_state=42)
    clf_tab.fit(X_train_sel, y_train)
    s_tab_train = clf_tab.predict_proba(X_train_sel)[:, 1]
    s_tab_test = clf_tab.predict_proba(X_test_sel)[:, 1]
    
    folds_data.append({
        's_tab_train': s_tab_train,
        's_tab_test': s_tab_test,
        's_soft_train': s_soft_all[train_idx],
        's_soft_test': s_soft_all[test_idx],
        's_hard_train': s_hard_all[train_idx],
        's_hard_test': s_hard_all[test_idx],
        'y_train': y_train
    })

best_c_params, best_b_params = precompute_best_mlp_params(folds_data, folds_data)
seeds = list(range(100))
results = []
print("Running 100 seeds for MLP baselines in parallel...")
max_workers = min(os.cpu_count() or 4, 8)
with ProcessPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(evaluate_mlp_seed, seed, folds_data, folds_data, best_c_params, best_b_params) for seed in seeds]
    for idx, future in enumerate(futures):
        results.append(future.result())

accs_concat, aucs_concat = [], []
accs_bilinear, aucs_bilinear = [], []
for r in results:
    p_c, pb_c = r['concat']
    p_b, pb_b = r['bilinear']
    accs_concat.append(accuracy_score(y_labels, p_c))
    aucs_concat.append(roc_auc_score(y_labels, pb_c))
    accs_bilinear.append(accuracy_score(y_labels, p_b))
    aucs_bilinear.append(roc_auc_score(y_labels, pb_b))

print("\n" + "=" * 96)
print("LOOCV MULTI-SEED BENCHMARK: NON-LINEAR MLP BASELINES WITH ACTIVATION SWEEPS (100 Seeds, N=46)")
print("=" * 96)
print(f"Late Concatenation MLP Acc: {np.mean(accs_concat)*100:.2f}% ± {np.std(accs_concat)*100:.2f}% (ROC-AUC: {np.mean(aucs_concat):.4f})")
print(f"Bilinear Tensor MLP Acc:    {np.mean(accs_bilinear)*100:.2f}% ± {np.std(accs_bilinear)*100:.2f}% (ROC-AUC: {np.mean(aucs_bilinear):.4f})")
print("=" * 96)
```

```python
# Cell 10: Multi-Seed Evaluation of CAF (Softplus scaling, N=46) with Implicit Questionnaire-based Gating (ANOVA K=16)
import os
import sys
import warnings
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
from concurrent.futures import ProcessPoolExecutor
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score, roc_auc_score
from scipy import stats
from src.data.dataset import ClinicalAttentionDataset

warnings.filterwarnings('ignore')

class CAFQuestionnaireGate(nn.Module):
    def __init__(self, gate_input_dim=16):
        super().__init__()
        self.gating = nn.Linear(gate_input_dim, 3)
        self.scale_param = nn.Parameter(torch.tensor([0.5413]))  # softplus(0.5413) approx 1.0
        self.classifier_bias = nn.Parameter(torch.zeros(1))
        
    def forward(self, x_gate_features, s_tab, s_soft, s_hard):
        gate_logits = self.gating(x_gate_features)
        probs = F.softmax(gate_logits, dim=1)  # [Batch, 3]
        p_tab = probs[:, 0:1]
        p_soft = probs[:, 1:2]
        p_hard = probs[:, 2:3]
        
        w_tab = p_tab * s_tab
        w_soft = p_soft * s_soft
        w_hard = p_hard * s_hard
        
        scale = F.softplus(self.scale_param)
        logit = scale * (w_tab + w_soft + w_hard) + self.classifier_bias
        return logit, probs

def train_and_eval_q_gate(folds_data, seed):
    torch.set_num_threads(1)
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    
    y_pred, y_prob = [], []
    attention_weights = []
    for fold in folds_data:
        s_tab_train, s_tab_test = fold['s_tab_train'], fold['s_tab_test']
        s_soft_train, s_soft_test = fold['s_soft_train'], fold['s_soft_test']
        s_hard_train, s_hard_test = fold['s_hard_train'], fold['s_hard_test']
        y_train = fold['y_train']
        X_gate_train, X_gate_test = fold['X_gate_train'], fold['X_gate_test']
        
        model = CAFQuestionnaireGate(gate_input_dim=16)
        criterion = nn.BCEWithLogitsLoss()
        optimizer = optim.LBFGS(
            model.parameters(), 
            lr=0.05, 
            max_iter=100,
            tolerance_grad=1e-7,
            tolerance_change=1e-9,
            history_size=50
        )
        
        s_tab_train_t = torch.tensor(s_tab_train, dtype=torch.float32).unsqueeze(1)
        s_soft_train_t = torch.tensor(s_soft_train, dtype=torch.float32).unsqueeze(1)
        s_hard_train_t = torch.tensor(s_hard_train, dtype=torch.float32).unsqueeze(1)
        y_train_t = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
        X_gate_train_t = torch.tensor(X_gate_train, dtype=torch.float32)
        X_gate_test_t = torch.tensor(X_gate_test, dtype=torch.float32)
        
        def closure():
            optimizer.zero_grad()
            logits, _ = model(X_gate_train_t, s_tab_train_t, s_soft_train_t, s_hard_train_t)
            loss = criterion(logits, y_train_t)
            loss.backward()
            return loss
            
        model.train()
        optimizer.step(closure)
        
        s_tab_test_t = torch.tensor(s_tab_test, dtype=torch.float32).unsqueeze(0)
        s_soft_test_t = torch.tensor(s_soft_test, dtype=torch.float32).unsqueeze(0)
        s_hard_test_t = torch.tensor(s_hard_test, dtype=torch.float32).unsqueeze(0)
        
        model.eval()
        with torch.no_grad():
            logit, probs = model(X_gate_test_t, s_tab_test_t, s_soft_test_t, s_hard_test_t)
            prob = torch.sigmoid(logit).item()
            y_pred.append(1 if prob >= 0.5 else 0)
            y_prob.append(prob)
            attention_weights.append(probs.numpy().flatten())
            
    return np.array(y_pred), np.array(y_prob), np.array(attention_weights)

project_root = r"c:\Users\abura\Development\CAF"
cleaned_xlsx_path = os.path.join(project_root, 'data', 'questionnaire_cleaned.xlsx')
dataset = ClinicalAttentionDataset(cleaned_xlsx_path)

exclude_ids = {6, 7, 31, 34}
keep_indices = [i for i, pid in enumerate(dataset.patient_ids) if pid not in exclude_ids]
dataset.patient_ids = [dataset.patient_ids[i] for i in keep_indices]
dataset.labels = [dataset.labels[i] for i in keep_indices]
dataset.X_gate = dataset.X_gate[keep_indices]
dataset.s_soft = dataset.s_soft[keep_indices]
dataset.s_hard = dataset.s_hard[keep_indices]
dataset.X_tab = dataset.X_tab[keep_indices]

y_labels = np.array(dataset.labels)
s_soft_all = dataset.s_soft
s_hard_all = dataset.s_hard
X_tab_all = dataset.X_tab
X_gate_all = dataset.X_gate

loo = LeaveOneOut()
folds_data = []
for train_idx, test_idx in loo.split(X_tab_all):
    X_train_tab, X_test_tab = X_tab_all[train_idx], X_tab_all[test_idx]
    y_train = y_labels[train_idx]
    
    selector = SelectKBest(score_func=f_classif, k=16)
    X_train_sel = selector.fit_transform(X_train_tab, y_train)
    X_test_sel = selector.transform(X_test_tab)
    
    clf_tab = LogisticRegression(C=1.0, solver='liblinear', random_state=42)
    clf_tab.fit(X_train_sel, y_train)
    s_tab_train = clf_tab.predict_proba(X_train_sel)[:, 1]
    s_tab_test = clf_tab.predict_proba(X_test_sel)[:, 1]
    
    folds_data.append({
        's_tab_train': s_tab_train,
        's_tab_test': s_tab_test,
        's_soft_train': s_soft_all[train_idx],
        's_soft_test': s_soft_all[test_idx],
        's_hard_train': s_hard_all[train_idx],
        's_hard_test': s_hard_all[test_idx],
        'y_train': y_train,
        'X_gate_train': X_train_sel,
        'X_gate_test': X_test_sel
    })

seeds = list(range(100))
results = []
print("Running 100 seeds for CAF with Questionnaire-based Gating...")
max_workers = min(os.cpu_count() or 4, 8)
with ProcessPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(train_and_eval_q_gate, folds_data, seed) for seed in seeds]
    for idx, future in enumerate(futures):
        results.append(future.result())

is_vaper = (X_gate_all[:, 2] > 0.0).astype(int)
vaper_indices = np.where(is_vaper == 1)[0]
non_vaper_indices = np.where(is_vaper == 0)[0]

accs, aucs, p_vals, p_soft_nv, p_soft_v = [], [], [], [], []
for r in results:
    pred, prob, weights = r
    accs.append(accuracy_score(y_labels, pred))
    aucs.append(roc_auc_score(y_labels, prob))
    
    p_soft_w = weights[:, 1]
    p_soft_nv.append(p_soft_w[non_vaper_indices].mean())
    p_soft_v.append(p_soft_w[vaper_indices].mean())
    
    _, p_val = stats.ttest_ind(p_soft_w[vaper_indices], p_soft_w[non_vaper_indices], equal_var=False)
    p_vals.append(p_val)

acc_m, acc_s = np.mean(accs), np.std(accs)
auc_m, auc_s = np.mean(aucs), np.std(aucs)
success = sum(1 for p in p_vals if p < 0.05) / 100 * 100
med_p = np.median(p_vals)
nv_mean = np.mean(p_soft_nv)
v_mean = np.mean(p_soft_v)

print("\n" + "=" * 96)
print("LOOCV MULTI-SEED EVALUATION: IMPLICIT QUESTIONNAIRE GATING (100 Seeds, N=46)")
print("=" * 96)
print(f"Accuracy:        {acc_m*100:.2f}% ± {acc_s*100:.2f}%")
print(f"ROC-AUC:         {auc_m:.4f} ± {auc_s:.4f}")
print(f"t-test Success:  {success:.1f}% (seeds with p < 0.05)")
print(f"Median p-value:  {med_p:.5e}")
print(f"Non-Vaper P_soft:{nv_mean:.4f}")
print(f"Vaper P_soft:    {v_mean:.4f}")
print("=" * 96)
```

```python
# Cell 11: Multi-Seed Evaluation of CAF (Softplus scaling, N=46) with Clinical Probing (s_hard) set to constant 1.0
import os
import sys
import warnings
import numpy as np
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import random
from concurrent.futures import ProcessPoolExecutor
from sklearn.linear_model import LogisticRegression
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.model_selection import LeaveOneOut
from sklearn.metrics import accuracy_score, roc_auc_score
from scipy import stats
from src.data.dataset import ClinicalAttentionDataset

warnings.filterwarnings('ignore')

class CAFSoftplus(nn.Module):
    def __init__(self):
        super().__init__()
        self.gating = nn.Linear(3, 3)
        self.scale_param = nn.Parameter(torch.tensor([0.5413]))  # softplus(0.5413) approx 1.0
        self.classifier_bias = nn.Parameter(torch.zeros(1))
        
    def forward(self, x_gate, s_tab, s_soft, s_hard):
        x_gate_vaping = x_gate[:, 2:5]
        gate_logits = self.gating(x_gate_vaping)
        probs = F.softmax(gate_logits, dim=1)  # [Batch, 3]
        p_tab = probs[:, 0:1]
        p_soft = probs[:, 1:2]
        p_hard = probs[:, 2:3]
        
        w_tab = p_tab * s_tab
        w_soft = p_soft * s_soft
        w_hard = p_hard * s_hard
        
        scale = F.softplus(self.scale_param)
        logit = scale * (w_tab + w_soft + w_hard) + self.classifier_bias
        return logit, probs

def evaluate_caf_seed(folds_data, seed):
    torch.set_num_threads(1)
    torch.manual_seed(seed)
    np.random.seed(seed)
    random.seed(seed)
    
    y_pred, y_prob = [], []
    attention_weights = []
    for fold in folds_data:
        s_tab_train, s_tab_test = fold['s_tab_train'], fold['s_tab_test']
        s_soft_train, s_soft_test = fold['s_soft_train'], fold['s_soft_test']
        s_hard_train, s_hard_test = fold['s_hard_train'], fold['s_hard_test']
        y_train = fold['y_train']
        X_gate_train, X_gate_test = fold['X_gate_train'], fold['X_gate_test']
        
        model = CAFSoftplus()
        criterion = nn.BCEWithLogitsLoss()
        optimizer = optim.LBFGS(
            model.parameters(), 
            lr=0.05, 
            max_iter=100,
            tolerance_grad=1e-7,
            tolerance_change=1e-9,
            history_size=50
        )
        
        s_tab_train_t = torch.tensor(s_tab_train, dtype=torch.float32).unsqueeze(1)
        s_soft_train_t = torch.tensor(s_soft_train, dtype=torch.float32).unsqueeze(1)
        s_hard_train_t = torch.tensor(s_hard_train, dtype=torch.float32).unsqueeze(1)
        y_train_t = torch.tensor(y_train, dtype=torch.float32).unsqueeze(1)
        X_gate_train_t = torch.tensor(X_gate_train, dtype=torch.float32)
        X_gate_test_t = torch.tensor(X_gate_test, dtype=torch.float32)
        
        def closure():
            optimizer.zero_grad()
            logits, _ = model(X_gate_train_t, s_tab_train_t, s_soft_train_t, s_hard_train_t)
            loss = criterion(logits, y_train_t)
            loss.backward()
            return loss
            
        model.train()
        optimizer.step(closure)
        
        s_tab_test_t = torch.tensor(s_tab_test, dtype=torch.float32).unsqueeze(0)
        s_soft_test_t = torch.tensor(s_soft_test, dtype=torch.float32).unsqueeze(0)
        s_hard_test_t = torch.tensor(s_hard_test, dtype=torch.float32).unsqueeze(0)
        
        model.eval()
        with torch.no_grad():
            logit, probs = model(X_gate_test_t, s_tab_test_t, s_soft_test_t, s_hard_test_t)
            prob = torch.sigmoid(logit).item()
            y_pred.append(1 if prob >= 0.5 else 0)
            y_prob.append(prob)
            attention_weights.append(probs.numpy().flatten())
            
    return np.array(y_pred), np.array(y_prob), np.array(attention_weights)

project_root = r"c:\Users\abura\Development\CAF"
cleaned_xlsx_path = os.path.join(project_root, 'data', 'questionnaire_cleaned.xlsx')
dataset = ClinicalAttentionDataset(cleaned_xlsx_path)

exclude_ids = {6, 7, 31, 34}
keep_indices = [i for i, pid in enumerate(dataset.patient_ids) if pid not in exclude_ids]
dataset.patient_ids = [dataset.patient_ids[i] for i in keep_indices]
dataset.labels = [dataset.labels[i] for i in keep_indices]
dataset.X_gate = dataset.X_gate[keep_indices]
dataset.s_soft = dataset.s_soft[keep_indices]
dataset.s_hard = dataset.s_hard[keep_indices] * 0.0 + 1.0
dataset.X_tab = dataset.X_tab[keep_indices]

y_labels = np.array(dataset.labels)
X_gate_all = dataset.X_gate
s_soft_all = dataset.s_soft
s_hard_all = dataset.s_hard
X_tab_all = dataset.X_tab

loo = LeaveOneOut()
folds_data = []
for train_idx, test_idx in loo.split(X_tab_all):
    X_train_tab, X_test_tab = X_tab_all[train_idx], X_tab_all[test_idx]
    y_train = y_labels[train_idx]
    
    selector = SelectKBest(score_func=f_classif, k=16)
    X_train_sel = selector.fit_transform(X_train_tab, y_train)
    X_test_sel = selector.transform(X_test_tab)
    
    clf_tab = LogisticRegression(C=1.0, solver='liblinear', random_state=42)
    clf_tab.fit(X_train_sel, y_train)
    s_tab_train = clf_tab.predict_proba(X_train_sel)[:, 1]
    s_tab_test = clf_tab.predict_proba(X_test_sel)[:, 1]
    
    folds_data.append({
        's_tab_train': s_tab_train,
        's_tab_test': s_tab_test,
        's_soft_train': s_soft_all[train_idx],
        's_soft_test': s_soft_all[test_idx],
        's_hard_train': s_hard_all[train_idx],
        's_hard_test': s_hard_all[test_idx],
        'y_train': y_train,
        'X_gate_train': X_gate_all[train_idx],
        'X_gate_test': X_gate_all[test_idx]
    })

seeds = list(range(100))
results = []
print("Running 100 seeds for constant 1.0 Probing CAF...")
max_workers = min(os.cpu_count() or 4, 8)
with ProcessPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(evaluate_caf_seed, folds_data, seed) for seed in seeds]
    for idx, future in enumerate(futures):
        results.append(future.result())

is_vaper = (X_gate_all[:, 2] > 0.0).astype(int)
vaper_indices = np.where(is_vaper == 1)[0]
non_vaper_indices = np.where(is_vaper == 0)[0]

accs, aucs, p_vals = [], [], []
p_tab_means, p_soft_means, p_hard_means = [], [], []
p_soft_nv, p_soft_v = [], []
for r in results:
    pred, prob, weights = r
    accs.append(accuracy_score(y_labels, pred))
    aucs.append(roc_auc_score(y_labels, prob))
    
    p_tab_means.append(weights[:, 0].mean())
    p_soft_means.append(weights[:, 1].mean())
    p_hard_means.append(weights[:, 2].mean())
    
    p_soft_w = weights[:, 1]
    p_soft_nv.append(p_soft_w[non_vaper_indices].mean())
    p_soft_v.append(p_soft_w[vaper_indices].mean())
    
    _, p_val = stats.ttest_ind(p_soft_w[vaper_indices], p_soft_w[non_vaper_indices], equal_var=False)
    p_vals.append(p_val)

acc_m, acc_s = np.mean(accs), np.std(accs)
auc_m, auc_s = np.mean(aucs), np.std(aucs)
success = sum(1 for p in p_vals if p < 0.05) / 100 * 100
med_p = np.median(p_vals)

print("\n" + "=" * 96)
print("LOOCV MULTI-SEED EVALUATION: CONSTANT 1.0 PROBING INPUT (100 Seeds, N=46)")
print("=" * 96)
print(f"Accuracy:        {acc_m*100:.2f}% ± {acc_s*100:.2f}%")
print(f"ROC-AUC:         {auc_m:.4f} ± {auc_s:.4f}")
print(f"Mean P_tab:      {np.mean(p_tab_means):.4f}")
print(f"Mean P_soft:     {np.mean(p_soft_means):.4f}")
print(f"Mean P_probing:  {np.mean(p_hard_means):.4f}")
print(f"t-test Success:  {success:.1f}% (seeds with p < 0.05)")
print(f"Median p-value:  {med_p:.5e}")
print(f"Non-Vaper P_soft:{np.mean(p_soft_nv):.4f}")
print(f"Vaper P_soft:    {np.mean(p_soft_v):.4f}")
print("=" * 96)
```
